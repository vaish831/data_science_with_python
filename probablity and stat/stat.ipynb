```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
from collections import Counter
import seaborn as sns
```

```python
sns.set(style="whitegrid")
```

```python
marks=[95,68,78,50,79,98,87,55,67,73,67]
df=pd.DataFrame({"marks":marks})
print(df)
```

### mean

```python
mean_value=np.mean(marks)
print(mean_value)
```

### median

```python
median_value=np.median(marks)
print(median_value)
```

### mode

```python
mode_value=Counter(marks).most_common(2)[0][0]
print(marks)
print(mode_value)
```

### variance

```python
variance_value=np.var(marks)
print(variance_value)
```

### standerd deviation

```python
standerd_value=np.std(marks)
print(marks)
print(standerd_value)
```

### histogram with matplotlib

```python
plt.figure(figsize = (8,5))
plt.hist(marks,color="blue",edgecolor="black",bins=5)
plt.title("histogram")
plt.xlabel("students marks")
plt.ylabel("no of students")
plt.show()
```

### histogram with seaborn

```python
plt.figure(figsize = (8,5))
sns.histplot(df["marks"],bins=5,kde=True,color="green")
plt.title("histogram")
plt.xlabel("students marks")
plt.ylabel("no of students")
plt.show()
```
